# SEABORN

Seaborn is a Python data visualization library built on top of Matplotlib and closely integrated with Pandas. It provides a high-level interface for creating attractive and informative statistical graphics.

Features of Seaborn:
Better Aesthetics: Produces more visually appealing plots than Matplotlib.
Built-in Themes: Comes with different styles and color palettes.
Statistical Plots: Supports plots like histograms, violin plots, box plots, and KDE plots.
Works Well with Pandas: Easily handles DataFrames and performs statistical aggregations.
Automatic Estimations: Handles missing values and estimates best-fit regression models.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load example dataset
var = sns.load_dataset("penguins")

# Set Seaborn style
sns.set(style="dark")

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Define custom order and hue order
order_1 = ["Dream", "Torgersen", "Biscoe"]
hue_order_1 = ["Female", "Male"]

# Create bar plot with all parameters
sns.barplot(
    x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, hue_order=hue_order_1,
    estimator=sum, ci=95, n_boot=1000, seed=42, units="species",
    orient="v", color=None, palette="coolwarm",
    saturation=0.8, errcolor="black", errwidth=2, capsize=0.2,
    dodge=True, width=0.7, ax=ax, alpha=0.8
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
data = np.random.rand(10, 10)  # 10x10 random matrix
df = pd.DataFrame(data, columns=[f"Col{i}" for i in range(10)])

# Create heatmap with all parameters
sns.heatmap(
    df,                         # Data
    vmin=0, vmax=1,             # Normalize color range
    cmap="coolwarm",            # Colormap
    center=0.5,                 # Centering color
    annot=True,                 # Show cell values
    fmt=".2f",                  # Format numbers
    annot_kws={"size": 8, "color": "black"},  # Text annotation properties
    linewidths=0.5,             # Line width between cells
    linecolor="gray",           # Line color
    cbar=True,                  # Show color bar
    cbar_kws={"shrink": 0.8, "orientation": "vertical"},  # Color bar customization
    xticklabels=True,           # Show x-axis labels
    yticklabels=True,           # Show y-axis labels
    mask=np.triu(np.ones_like(df, dtype=bool)),  # Mask upper triangle
    square=True,                # Square cells
    robust=True,                # Avoid extreme outliers
)

plt.title("Seaborn Heatmap with All Parameters", fontsize=14)
plt.xlabel("Columns", fontsize=12)
plt.ylabel("Rows", fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels
plt.yticks(rotation=0)   # Keep y labels horizontal

plt.show()
```

# Factor Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
var = sns.load_dataset("tips")

# Use catplot instead of factorplot
sns.catplot(x="size", y="tip", data=var, kind="point")  # 'tip' instead of 'tips'

# Show the plot
plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var,hue="sex",kind="point")  
plt.show()
```

 1. Strip Plot (kind="strip")
Used for: Showing individual data points

```python
sns.catplot(x="day", y="total_bill", data=var, kind="strip", jitter=True)
plt.show()
```

 2. Swarm Plot (kind="swarm")
Used for: Avoiding overlapping in a scatter plot

```python

sns.catplot(x="day", y="total_bill", data=var, kind="swarm")
plt.show()
```

 3. Box Plot (kind="box")
Used for: Showing data distribution with quartiles and outliers

```python
sns.catplot(x="day", y="total_bill", data=var, kind="box")
plt.show()
```

 4. Violin Plot (kind="violin")
Used for: Showing distribution with KDE (density estimation)

```python
sns.catplot(x="day", y="total_bill", data=var, kind="violin", split=True, hue="sex")
plt.show()
```

 5. Boxen Plot (kind="boxen")
Used for: Better visualization of large datasets

```python
sns.catplot(x="day", y="total_bill", data=var, kind="boxen")
plt.show()
```

 6. Bar Plot (kind="bar")
Used for: Showing mean values with confidence intervals

```python
sns.catplot(x="day", y="total_bill", data=var, kind="bar", hue="sex")
plt.show()
```

 7. Count Plot (kind="count")
Used for: Counting occurrences of categories

```python
sns.catplot(x="day", data=var, kind="count", hue="sex")
plt.show()
```

 8. Point Plot (kind="point")
Used for: Showing trends across categories

```python
sns.catplot(x="day", y="total_bill", data=var, kind="point", hue="sex",height=10)
plt.show()
```
